```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])
df['job_type_skills'] = df['job_type_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else {})
```

```python
df_da = df[df['job_title_short'] == 'Data Analyst'].copy()
df_da
```

```python
# Extract the month from the job_posted_date
df_da['job_posted_month_no'] = df_da['job_posted_date'].dt.month
```

```python
df_da_explode = df_da.explode('job_skills')
```

```python
# Lets pivot the data to get the count of each skill by month
df_da_pivot = df_da_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_da_pivot
```

```python
# we need to sort the skills to be in order of the count to make it easier to plot
# create a total row
df_da_pivot.loc['Total'] = df_da_pivot.sum()
# order columns by total count
df_da_pivot = df_da_pivot[df_da_pivot.loc['Total'].sort_values(ascending=False).index]
# drop total row
df_da_pivot = df_da_pivot.drop('Total')

df_da_pivot
```

```python
# I like to have my months as month names instead of numbers, so I will create a mapping of month numbers to month names
month_mapping = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
                 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
df_da_pivot.index = df_da_pivot.index.map(month_mapping)
df_da_pivot
```

```python
# Another way of adding the month names is to use the calendar module
import calendar

month_order = list(calendar.month_name)[1:]  # Jan–Dec

df_da_pivot = df_da_pivot.reindex(month_order)
df_da_pivot
```

```python
# Lets graph the top 5 skills by month
top_5_skills = df_da_pivot.sum().sort_values(ascending=False).head(5).index
df_da_pivot[top_5_skills].plot(kind='line', marker='o')
plt.title('Top 5 Skills for Data Analysts by Month')
plt.xlabel('Month')
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45)
plt.legend(title='Skills')
plt.grid()
plt.tight_layout()
plt.show()
```
