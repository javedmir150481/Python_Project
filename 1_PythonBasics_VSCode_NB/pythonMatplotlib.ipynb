```python
# Install datasets Library (if not already installed)
!pip install -U datasets

# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  # we will use this for plotting

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
x = [1,3,2,5]
y = [2,1,0,5]
plt.plot(x,y)
plt.show()
```

```python
date_counts = df["job_posted_date"].value_counts()
date_counts = date_counts.sort_index()
print(date_counts)
```

```python
plt.plot(date_counts.index, date_counts.values)
```

```python
# new column
df["job_posted_month"] = df["job_posted_date"].dt.month
```

```python
job_counts = df["job_title_short"].value_counts()
job_counts = job_counts.sort_index()
job_counts.sort_values(ascending=True, inplace=True)
plt.barh(job_counts.index, job_counts.values)
```
