```python
import numpy as np
```

```python
job_titles = np.array(["Data Analyst", "Data Scientist", "Data Engineer", "Machine Learning Engineer"])
base_salaries = np.array([60000, 80000, 75000, 90000])
bonus_rates = np.array([.05, .1, .08, .12])
```

```python
np.mean(total_salaries)
```
