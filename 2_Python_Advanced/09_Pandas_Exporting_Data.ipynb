```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.head().to_csv("jobs_data.csv", index=False)
```

```python
df.head().to_excel("jobs_data.xlsx", index=False)
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///jobs_data.db')

df.head().to_sql("job_table", con=engine, if_exists="append", index=False)
```

```python
df.head().to_parquet("data_jobs.parquet", index=False)
```

```python
df.head().to_pickle("data_jobs.pkl")
```
