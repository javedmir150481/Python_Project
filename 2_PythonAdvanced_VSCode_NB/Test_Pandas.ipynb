```python
df.groupby(['job_country', 'job_title_short'])['salary_year_avg'].agg(['max', 'median', 'min']).dropna()
```

```python
df.pivot_table(
    values='salary_year_avg',
    index=['job_country', 'job_title_short'],
    aggfunc={'salary_year_avg': ['min', 'max', 'median']}
)
```

```python
df_job_country_salary = df.pivot_table(
    values='salary_year_avg',        # The data to aggregate
    index='job_country',             # Rows (index of pivot table)
    columns='job_title_short',       # Columns (pivot table headers)
    aggfunc='median'                 # Aggregation function
)

df_job_country_salary
```

```python
top_countries = df['job_country'].value_counts().head(5).index

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = df['job_title_short'].value_counts().head(5).index
df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind='bar', figsize=(12, 6))
plt.title('Median Salary by Job Country and Job Title')
plt.xlabel('Job Country')
plt.ylabel('Median Salary (Yearly)')
plt.xticks(rotation=45)
plt.legend(title='Job Title')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

```python
df_india = df[df['job_country'] == 'India'].copy()
df_india.head()
```

```python
df_india["job_posted_month"] = df_india["job_posted_date"].dt.strftime("%B")
df_india.head()
```

```python
df_india_pivot = df_india.pivot_table(index='job_posted_month', columns="job_title_short", aggfunc='size')
df_india_pivot
```

```python
# Long way to sort my months in the right order. I will show you a better way in the next section.
df_india_pivot = df_india_pivot.reset_index()
df_india_pivot["job_posted_month_no"] = pd.to_datetime(df_india_pivot["job_posted_month"], format="%B").dt.month
df_india_pivot.sort_values("job_posted_month_no", inplace=True)
df_india_pivot.set_index("job_posted_month", inplace=True)
df_india_pivot.drop(columns=["job_posted_month_no"], inplace=True)
df_india_pivot
```

```python
df_india_pivot.plot(kind='line', figsize=(12, 6))
```

```python
top_3_jobs_india = df_india['job_title_short'].value_counts().head(3).index.to_list()
df_india_pivot[top_3_jobs_india].plot(kind='line', figsize=(8, 6))
```

```python
# short way to sort my months in the right order.
df_india["job_posted_month"] = df_india["job_posted_date"].dt.to_period("M")

df_india_pivot_short = df_india.pivot_table(index='job_posted_month', columns="job_title_short", aggfunc='size')
df_india_pivot_short
```

```python
df_india_pivot_short.plot(kind='line', figsize=(8, 6))
plt.title('Job Postings Over Time in India')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.xticks(rotation=45)
plt.legend(title='Job Title')
plt.tight_layout()
plt.show()
```
